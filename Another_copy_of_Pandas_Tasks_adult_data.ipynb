Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
data2 = pd.DataFrame(data,columns=["age","workclass","fnlwgt","education","educational-num","marital-status","occupation","relationship","race","gender","capital-gain","capital-loss","hours-per-week","native-country","income"])
data2
data2['income_binary']=data2['income'].apply(lambda x: 0 if x=='<=50k'else 1)
data2
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
data2=pd.DataFrame(data)
```

```python
def categorize_age(age):
    if age < 30:
        return 'Youth'
    elif age >= 30 and age < 60:
        return 'Adult'
    else:
        return 'Senior'
```

```python
data2['age_group']=data2['age'].apply(categorize_age)
```
